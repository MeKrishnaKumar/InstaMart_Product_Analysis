**Visualization**

```python
df.plot()
```

```python
df.plot(x='Price', y='Rating', kind='scatter')
plt.title('Comparision between Price and Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.set(style="whitegrid", font_scale=1.2, rc={"figure.figsize":(20,10)})
sns.barplot(data=df,x = 'Name', y = 'Rating',color='green',saturation=0.35)
plt.xticks(rotation=90)
plt.tight_layout()
plt.xlabel('Name',fontsize=30)
plt.ylabel('Rating',fontsize=30)
```

```python
import plotly.graph_objs as go

# Create a trace
trace = go.Pie(labels=df['Name'], values=df['Price'])

# Create a layout
layout = go.Layout(title='Product Distributions')

# Create a figure
fig = go.Figure(data=[trace], layout=layout)

# Show the figure
fig.show()
```

```python
# Create a trace
trace = go.Pie(labels=df['Company'], values=df['Price'])

# Create a layout
layout = go.Layout(title='Company Distributions')

# Create a figure
fig = go.Figure(data=[trace], layout=layout)

# Show the figure
fig.show()
```

```python
# Show the plot
sns.countplot(x ='Keyword', data = df)
plt.show()
```

```python
sns.set(style="whitegrid", font_scale=1.2, rc={"figure.figsize":(20,10)})
sns.barplot(data=df,x = 'Keyword', y = 'Rating',color='green',saturation=0.35)
plt.xticks(rotation=90)
plt.tight_layout()
plt.xlabel('Keyword',fontsize=30)
plt.ylabel('Rating',fontsize=30)
```
